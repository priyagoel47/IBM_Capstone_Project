Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

QUERY = """SELECT * FROM master"""

df = pd.read_sql_query(QUERY,conn)

plt.hist(df['ConvertedComp'])
plt.title('Histogram of ConvertedComp')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
df['Age'].plot(kind='box', figsize=(8, 6))
plt.title('Boxplot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(8,6))
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
df.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', s='Age', figsize=(8,6), alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs based on Age')
plt.legend(['Age'], loc='upper right')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY1 = """SELECT * FROM DatabaseDesireNextYear"""

df1 = pd.read_sql_query(QUERY1,conn)
df_group = df1.groupby('DatabaseDesireNextYear', axis=0).count()
df_group.sort_values(['Respondent'], ascending=False, inplace=True)
df_group.reset_index(inplace=True)
df_new = df_group.head(5)
plt.pie(df_new['Respondent'], labels=df_new['DatabaseDesireNextYear'], autopct = '%1.1f%%')
plt.title('Pie Chart of Top 5 Databases')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
df_chk = df[(df['Age']>=30.0) & (df['Age'] <= 35.0)]
df_fin = df_chk[['WorkWeekHrs','CodeRevHrs', 'Age']].groupby('Age', axis=0).median()
df_fin
df_fin.plot(kind = 'bar', stacked=True, figsize=(10,6))
plt.title('Stacked Chart of WorkWeekHrs and CodeRevHrs For the Age Group 30-35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df_age = df[(df['Age']>=45.0) & (df['Age'] <= 60.0)]
df_final = df_age[['ConvertedComp', 'Age']].groupby('Age', axis=0).median()
df_final.plot(kind = 'line', figsize=(10,6))
plt.title('Line Chart of ConvertedComp For the Age Group 45-60')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
df['MainBranch'].value_counts().plot(kind='barh', figsize=(10,6))
plt.show()
```
